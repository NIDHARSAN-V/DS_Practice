## 4. Hypothesis Testing

Apply statistical tests to evaluate relationships between features and readmission.

```python
df.columns = df.columns.str.strip()

df["readmitted_30days"] = (
    df["readmitted_30days"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": 1, "no": 0})
)

df = df[df["readmitted_30days"].notna()]

print("Target value counts:")
print(df["readmitted_30days"].value_counts())
```

```python
Q1 = df['length_of_stay'].quantile(0.25)
Q3 = df['length_of_stay'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

lower_outliers = df[df['length_of_stay'] < lower]
upper_outliers = df[df['length_of_stay'] > upper]
normal_data = df[
    (df['length_of_stay'] >= lower) &
    (df['length_of_stay'] <= upper)
]

print("--- Group Sizes ---")
print("Lower:", len(lower_outliers))
print("Upper:", len(upper_outliers))
print("Normal:", len(normal_data))
```

```python
def safe_mean(series):
    return round(series.mean(), 3) if len(series) > 0 else "No data"

mean_lower = safe_mean(lower_outliers['readmitted_30days'])
mean_upper = safe_mean(upper_outliers['readmitted_30days'])
mean_normal = safe_mean(normal_data['readmitted_30days'])

print("\n--- Readmission Rates ---")
print("Lower:", mean_lower)
print("Upper:", mean_upper)
print("Normal:", mean_normal)

if mean_upper > mean_normal:
    print("\n✅ Upper outliers have higher readmission risk")
```

```python
from scipy.stats import mannwhitneyu

normal = normal_data['readmitted_30days']
upper_grp = upper_outliers['readmitted_30days']

print("\n--- After Cleaning ---")
print("Normal size:", len(normal))
print("Upper size:", len(upper_grp))

if len(normal) == 0 or len(upper_grp) == 0:
    print("\n❌ Cannot perform test (empty group)")
else:
    stat, p_value = mannwhitneyu(normal, upper_grp, alternative='less')

    print("\n=== Hypothesis Test Result ===")
    print("P-value:", p_value)

    if p_value < 0.05:
        print("✅ Reject H0 → Significant difference")
    else:
        print("❌ Fail to reject H0 → Not significant")
```

```python
print("Conclusion:")
print("There is a statistically significant difference in readmission")
print("between normal and high length_of_stay patients.")
```

```python
from scipy.stats import mannwhitneyu
import pandas as pd

results = []

with open("full_outlier_analysis.txt", "w", encoding="utf-8") as f:

    def log(text):
        print(text)
        f.write(text + "\n")

    numeric_cols = df.select_dtypes(include=['number']).columns

    for col in numeric_cols:

        log("\n==============================")
        log(f"Column: {col}")
        log("==============================")

        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        lower_outliers = df[df[col] < lower]
        upper_outliers = df[df[col] > upper]
        normal_data = df[(df[col] >= lower) & (df[col] <= upper)]

        def safe_mean(series):
            return round(series.mean(), 3) if len(series) > 0 else "No data"

        mean_lower = safe_mean(lower_outliers['readmitted_30days'])
        mean_upper = safe_mean(upper_outliers['readmitted_30days'])
        mean_normal = safe_mean(normal_data['readmitted_30days'])

        log("\n--- Group Sizes ---")
        log(f"Lower: {len(lower_outliers)}")
        log(f"Upper: {len(upper_outliers)}")
        log(f"Normal: {len(normal_data)}")

        log("\n--- Readmission Rates ---")
        log(f"Lower: {mean_lower}")
        log(f"Upper: {mean_upper}")
        log(f"Normal: {mean_normal}")

        normal = normal_data['readmitted_30days']
        upper_grp = upper_outliers['readmitted_30days']
        lower_grp = lower_outliers['readmitted_30days']

        if len(normal) > 0 and len(upper_grp) > 0:
            _, p1 = mannwhitneyu(normal, upper_grp, alternative='less')
        else:
            p1 = None

        if len(normal) > 0 and len(lower_grp) > 0:
            _, p2 = mannwhitneyu(lower_grp, normal, alternative='less')
        else:
            p2 = None

        if len(lower_grp) > 0 and len(upper_grp) > 0:
            _, p3 = mannwhitneyu(lower_grp, upper_grp)
        else:
            p3 = None

        results.append({
            "feature": col,
            "p_normal_vs_upper": p1,
            "p_normal_vs_lower": p2,
            "p_lower_vs_upper": p3,
            "mean_lower": mean_lower,
            "mean_upper": mean_upper,
            "mean_normal": mean_normal
        })

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by="p_normal_vs_upper", na_position='last')

results_df.to_csv("summary_results.csv", index=False)

print("\n✅ Full analysis complete!")
```

```python
table = pd.crosstab(df['disease_count'], df['readmitted_30days'])

print("========== CONTINGENCY TABLE ==========")
print(table)
```

```python
print("H0: Disease count does NOT affect readmission")
print("H1: Disease count affects readmission")
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)

print("========== CHI-SQUARE RESULT ==========")
print("Chi2 value:", chi2)
print("P-value:", p)
```

```python
alpha = 0.05

print("========== DECISION ==========")

if p < alpha:
    print("✅ Reject H0 → Significant relationship")
else:
    print("❌ Fail to Reject H0 → No significant relationship")
```

```python
rate = table.div(table.sum(axis=1), axis=0)

print("========== READMISSION RATES ==========")
print(rate)
```

```python
if p < alpha:
    print("There is a statistically significant relationship between disease_count and readmission.")
    print("👉 As disease_count increases, readmission rate increases.")
else:
    print("No statistically significant relationship found.")
```

```python
cat_cols = [
    'has_diabetes',
]

target = 'readmitted_30days'

print("Selected features:", cat_cols)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

results = []

for col in cat_cols:

    print("\n" + "="*50)
    print(f"Column: {col}")
    print("="*50)

    table = pd.crosstab(df[col], df[target])

    print("\n--- Contingency Table ---")
    print(table)

    chi2, p, dof, expected = chi2_contingency(table)

    print("\nChi2 value:", chi2)
    print("P-value:", p)

    if p < 0.05:
        print("Result: ✅ Significant (Reject H0)")
        significance = "Significant"
    else:
        print("Result: ❌ Not Significant (Fail to Reject H0)")
        significance = "Not Significant"

    print("\n--- Readmission Rates ---")
    rate = table.div(table.sum(axis=1), axis=0)
    print(rate)

    results.append({
        "feature": col,
        "p_value": p,
        "chi2": chi2,
        "result": significance
    })
```

```python
results_df = pd.DataFrame(results)

print("\n========== FINAL SUMMARY ==========")
print(results_df.sort_values(by="p_value"))

results_df.to_csv("chi_square_results.csv", index=False)

print("✅ Results saved as chi_square_results.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_numeric = df.select_dtypes(include='number')

corr_matrix = df_numeric.corr()

plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

```python
import numpy as np

upper_triangle = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

to_drop = [
    column for column in upper_triangle.columns
    if any(abs(upper_triangle[column]) > 0.85)
]

print("⚠️ Highly correlated features (threshold > 0.4):")
print(to_drop)
```

```python
df = pd.read_csv("../dataset/base_raw.csv")

print("Dataset shape:", df.shape)
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')

def bucket_age(age):
    if age <= 35:
        return "low"
    elif age <= 60:
        return "medium"
    else:
        return "high"

df['age'] = df['age'].apply(bucket_age)

print(df['age'].value_counts())
```

```python
df['admission_type'] = df['admission_type'].str.strip().str.lower()
```

```python
df['admission_source'] = df['admission_source'].str.strip().str.lower()
```

```python
df['length_of_stay'] = pd.to_numeric(df['length_of_stay'], errors='coerce')

def bucket_stay(days):
    if days <= 3:
        return "short"
    elif days <= 7:
        return "medium"
    else:
        return "long"

df['length_of_stay'] = df['length_of_stay'].apply(bucket_stay)

print(df['length_of_stay'].value_counts())
```

```python
df['department'] = df['department'].str.strip().str.lower()
```

```python
df.drop('room_type', axis=1, inplace=True)

print("room_type dropped")
```

```python
def bucket_procedures(num):
    if num == 0:
        return "none"
    elif num <= 2:
        return "few"
    else:
        return "many"

df['num_procedures'] = df['num_procedures'].apply(bucket_procedures)

print(df['num_procedures'].value_counts())
```

```python
def bucket_medications(num):
    if num < 2:
        return "few"
    elif num <= 6:
        return "many"
    else:
        return "toomany"

df['num_medications'] = df['num_medications'].apply(bucket_medications)

print(df['num_medications'].value_counts())
```

```python
df['previous_admissions'] = pd.to_numeric(df['previous_admissions'], errors='coerce')

def bucket_prev_admissions(num):
    if num == 0:
        return "none"
    elif num <= 3:
        return "few"
    else:
        return "many"

df['previous_admissions'] = df['previous_admissions'].apply(bucket_prev_admissions)

print(df['previous_admissions'].value_counts())
```

### Derived Medical Features

```python
df['MAP'] = (
    df['discharge_systolic_bp'] +
    2 * df['discharge_diastolic_bp']
) / 3

df['pulse_pressure'] = (
    df['discharge_systolic_bp'] -
    df['discharge_diastolic_bp']
)

df['shock_index'] = (
    df['discharge_heart_rate'] /
    df['discharge_systolic_bp']
)

print("✅ Derived features created: MAP, pulse_pressure, shock_index")
```
